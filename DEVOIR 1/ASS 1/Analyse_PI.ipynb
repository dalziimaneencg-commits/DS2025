```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
student_performance = fetch_ucirepo(id=320)

# data (as pandas dataframes)
X = student_performance.data.features
y = student_performance.data.targets

# metadata
print(student_performance.metadata)

# variable information
print(student_performance.variables)
```

Let's first inspect the `X` (features) and `y` (targets) dataframes to understand their content and structure, which will help in deciding what to plot.

```python
print('Features (X) head:')
display(X.head())
print('\nFeatures (X) info:')
X.info()

print('\nTargets (y) head:')
display(y.head())
print('\nTargets (y) info:')
y.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Plotting histograms for G1, G2, G3
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(y['G1'], bins=20, kde=True)
plt.title('Distribution of G1 (First Period Grade)')
plt.xlabel('Grade')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
sns.histplot(y['G2'], bins=20, kde=True)
plt.title('Distribution of G2 (Second Period Grade)')
plt.xlabel('Grade')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
sns.histplot(y['G3'], bins=20, kde=True)
plt.title('Distribution of G3 (Final Grade)')
plt.xlabel('Grade')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(x=y['G1'], y=y['G3'])
plt.title('G1 vs G3')
plt.xlabel('G1 (First Period Grade)')
plt.ylabel('G3 (Final Grade)')

plt.subplot(1, 2, 2)
sns.scatterplot(x=y['G2'], y=y['G3'])
plt.title('G2 vs G3')
plt.xlabel('G2 (Second Period Grade)')
plt.ylabel('G3 (Final Grade)')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=X['sex'], y=y['G3'])
plt.title('Final Grade (G3) by Sex')
plt.xlabel('Sex')
plt.ylabel('G3 (Final Grade)')
plt.show()
```
